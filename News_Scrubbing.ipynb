```
def extract_values(obj, key):
    """Pull all values of specified key from nested JSON."""
    arr = []

    def extract(obj, arr, key):
        """Recursively search for values of key in JSON tree."""
        if isinstance(obj, dict):
            for k, v in obj.items():
                if isinstance(v, (dict, list)):
                    extract(v, arr, key)
                elif k == key:
                    arr.append(v)
        elif isinstance(obj, list):
            for item in obj:
                extract(item, arr, key)
        return arr

    results = extract(obj, arr, key)
    return results
```

```
import requests
import pandas as pd

from requests import Request, Session
from requests.exceptions import ConnectionError, Timeout, TooManyRedirects
import json

df = pd.DataFrame()

keys = ["id","name","author","title","description","url","urlToImage","publishedAt","content"]
 

url = ('http://newsapi.org/v2/top-headlines?'
       'country=in&'
       'apiKey=a39aeb55141d4016a7c77966f2a1a504') 
response = requests.get(url)


try:
  data = json.loads(response.text)
  count = extract_values(data, 'totalResults')[0]
  print("Stroies count : ", count/2)
  #print(data)
  value=[] 
  for i in range(len(keys)):
    value.append(extract_values(data, keys[i]))

  for j in range(count//2 +1):
   print("Printing News : ", j, "\n")
   for i in range(len(keys)):
    print(keys[i] , ": ", value[i][j])
    
    print("\n")
except (ConnectionError, Timeout, TooManyRedirects) as e:
  print(e)
```
